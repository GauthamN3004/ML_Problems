```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train_loan.csv")
df.head()
```

```python
df.shape
```

```python
df["Credit_History"].unique()
```

```python
df.isnull().sum()
```

```python
#filling missing values of gender
df.loc[df["Gender"].isnull(),"Gender"] = df['Gender'].value_counts().idxmax()
df["Gender"].unique()
```

```python
#filling missing values of married
df.loc[df["Married"].isnull(),"Married"] = df['Married'].value_counts().idxmax()
df["Married"].unique()
```

```python
#filling missing values of Dependents
df.loc[df["Dependents"].isnull(),"Dependents"] = df['Dependents'].value_counts().idxmax()
df["Dependents"].unique()
```

```python
#filling missing values of Self_Employed
df.loc[df["Self_Employed"].isnull(),"Self_Employed"] = df['Self_Employed'].value_counts().idxmax()
df["Self_Employed"].value_counts()
```

```python
#filling missing values of Self_Employed
df.loc[df["LoanAmount"].isnull(),"LoanAmount"] = df["LoanAmount"].mean()
df["LoanAmount"].mean()
```

```python
#filling missing values of Loan_Amount_Term
df.loc[df["Loan_Amount_Term"].isnull(),"Loan_Amount_Term"] = df["Loan_Amount_Term"].value_counts().idxmax()
df["Loan_Amount_Term"].unique()
```

```python
#filling missing values of Credit_History
df.loc[df["Credit_History"].isnull(),"Credit_History"] = df["Credit_History"].value_counts().idxmax()
df["Credit_History"].unique()
```

```python
df.isnull().sum()
```

```python
#drop column Loan_ID
df = df.iloc[:,1:]
df.head()
```

```python
df.loc[df["Gender"] == "Male","Gender"] = 1
df.loc[df["Gender"] == "Female","Gender"] = 0
df.head()
```

```python
df.loc[df["Married"] == "Yes","Married"] = 1
df.loc[df["Married"] == "No","Married"] = 0
df.head()
```

```python
df.loc[df["Education"] == "Graduate","Education"] = 1
df.loc[df["Education"] == "Not Graduate","Education"] = 0
df.head()
```

```python
df.loc[df["Self_Employed"] == "Yes","Self_Employed"] = 1
df.loc[df["Self_Employed"] == "No","Self_Employed"] = 0
df.head()
```

```python
#getting dummy variables
```

```python
pd.concat([df,pd.get_dummies(df["Dependents"]).iloc[:,:-1]],axis=1)
```

```python
df = pd.concat([df,pd.get_dummies(df["Dependents"]).iloc[:,:-1]],axis=1)
df.drop(columns = ["Dependents"],inplace = True)
df.head()
```

```python
df = pd.concat([df,pd.get_dummies(df["Loan_Amount_Term"]).iloc[:,:-1]],axis=1)
df.drop(columns = ["Loan_Amount_Term"],inplace = True)
df.head()
```

```python
#we see that column 6.0 and 350.0 is missing so we can make a new column with all zeros
df.insert(21,"350.0",np.zeros(df.shape[0]))
df.insert(13,"6.0",np.zeros(df.shape[0]))
df["350.0"] = df["350.0"].astype(int)
df.head()
```

```python
df.columns
```

```python
df["Property_Area"].unique()
```

```python
df = pd.concat([df,pd.get_dummies(df["Property_Area"]).iloc[:,1:]],axis=1)
df.drop(columns = ["Property_Area"],inplace = True)
df.head()
```

```python
df.loc[df["Loan_Status"]=='Y',"Loan_Status"] = 1
df.loc[df["Loan_Status"]=='N',"Loan_Status"] = 0
```

```python
X = df.iloc[:,:]
Y = df["Loan_Status"].values
X.drop(columns = ["Loan_Status"],inplace = True)
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.asarray(X)
Y = Y.astype(int)
Y = np.asarray(Y)
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size = 0.1,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import jaccard_similarity_score as jss
```

```python
knn = KNeighborsClassifier()
```

```python
parameters = [{'n_neighbors':[11,12,13,14]}]
```

```python
knn_gs = GridSearchCV(estimator = knn,param_grid = parameters,scoring = 'accuracy')
```

```python
knn_gs.fit(xtrain,ytrain)
```

```python
knn_gs.best_params_
```

```python
ypred = knn_gs.predict(xtest)
```

```python
knn_jss = jss(ytest,ypred)
knn_jss
```
