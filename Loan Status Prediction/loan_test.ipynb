```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("test_loan.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#filling missing values of gender
df.loc[df["Gender"].isnull(),"Gender"] = df['Gender'].value_counts().idxmax()
df["Gender"].value_counts()
```

```python
#filling missing values of Dependents
df.loc[df["Dependents"].isnull(),"Dependents"] = df['Dependents'].value_counts().idxmax()
df["Dependents"].value_counts()
```

```python
#filling missing values of Credit_History
df.loc[df["Credit_History"].isnull(),"Credit_History"] = df["Credit_History"].value_counts().idxmax()
df["Credit_History"].value_counts()
```

```python
#filling missing values of Self_Employed
df.loc[df["Self_Employed"].isnull(),"Self_Employed"] = df["Self_Employed"].value_counts().idxmax()
df["Self_Employed"].value_counts()
```

```python
df.loc[df["Gender"] == "Male","Gender"] = 1
df.loc[df["Gender"] == "Female","Gender"] = 0
df.head()
```

```python
df.loc[df["Married"] == "Yes","Married"] = 1
df.loc[df["Married"] == "No","Married"] = 0
df.head()
```

```python
df.loc[df["Self_Employed"] == "Yes","Self_Employed"] = 1
df.loc[df["Self_Employed"] == "No","Self_Employed"] = 0
df.head()
```

```python
#drop column Loan_ID
df = df.iloc[:,1:]
df.head()
```

```python
df["Dependents"].unique()
```

```python
pd.get_dummies(df["Dependents"]).iloc[:,:-1]
```

```python
df.loc[df["Education"] == "Graduate","Education"] = 1
df.loc[df["Education"] == "Not Graduate","Education"] = 0
df.head()
```

```python
df = pd.concat([df,pd.get_dummies(df["Dependents"]).iloc[:,:-1]],axis=1)
df.drop(columns = ["Dependents"],inplace = True)
df.head()
```

```python
df = pd.concat([df,pd.get_dummies(df["Loan_Amount_Term"]).iloc[:,:-1]],axis=1)
df.drop(columns = ["Loan_Amount_Term"],inplace = True)
df.head()
```

```python
df.columns
```

```python
df["Property_Area"].unique()
```

```python
df = pd.concat([df,pd.get_dummies(df["Property_Area"]).iloc[:,1:]],axis=1)
df.drop(columns = ["Property_Area"],inplace = True)
df.head()
```

```python
df.to_csv("mod_test_loan.csv",index=False)
df.shape
```
